```python
import numpy as np  
import pandas as pd
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B', 'C', 'A'],
    'Store': ['X', 'Y', 'X', 'Y', 'X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550],
    'Quantity': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D')
}

df = pd.DataFrame(data)
df
```

```python
cat = df.groupby('Category')
for name, group in cat:
    print(f"Category: {name}")
    print(group)
    print("\n")
```

```python
# Group by 'Category' and calculate total sales 
total_sales = df.groupby('Category')['Sales'].sum()
total_sales
```

```python
# Group by 'Category' and calculate total sales and average quantity
grouped = df.groupby('Category').agg({
    'Sales': 'sum',
    'Quantity': 'mean'
})

grouped
```

```python
# group by multiple columns (Category and Store) and calculate total sales
grouped_multi = df.groupby(['Category', 'Store'])['Sales'].sum()
grouped_multi
```
