```python
import numpy as np 
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

df1
```

```python
df1.shape  # returns a tuple of (number of rows, number of columns)
```

```python
df1.columns # returns the column labels of the DataFrame
```

```python
df1.info() # provides a concise summary of the DataFrame, including data types and non-null counts
```

```python
df1.describe() # generates descriptive statistics of the DataFrame, such as mean, std, min, max, etc. for numeric columns
```

```python
df1['A'] + 10 # adds 10 to each element in column 'A' and returns a new Series
```

```python
# defining a custom function to apply to each element in column 'B'
def square(x):
    return x ** 2
```

```python
df1['B'].apply(square) # applies the 'square' function to each element in column 'B' and returns a new Series (dont change the original DataFrame)
```

```python
df1['B'] = df1['B'].apply(square) # applies the 'square' function to each element in column 'B' and updates the original DataFrame with the squared values
```

```python
df1
```

```python
df1['D'] = df1['A'] * df1['B'] # creates a new column 'D' by multiplying columns 'A' and 'B' element-wise
```
