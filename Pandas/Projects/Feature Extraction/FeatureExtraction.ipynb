```python
import numpy as np  
import pandas as pd
```

```python
df = pd.read_csv(r'anime.csv')
df.head()
```

```python
df.loc[0]
```

```python
def extract_episodes(txt):
    if pd.isna(txt):
        return np.nan
    try:
        start = txt.find('(')
        end = txt.find(')')

        if start != -1 and end != -1 and end > start:
            inside_text = txt[start + 1:end]
            parts = inside_text.split()
        return int(parts[0])  # Extract the number of episodes
    except (ValueError, IndexError):
        return np.nan  # Return NaN if conversion fails
```

```python
episode = extract_episodes(df.loc[0]['Title'])
episode
```

### Make a new Column for episode count

```python
# extract the episode number for all the rows in the DataFrame and create a new column 'Episodes' to store these values
df['Episodes'] = df['Title'].apply(extract_episodes)
```

```python
def extract_time(title):
    if pd.isna(title):
        return np.nan
    try:
        part = title.split(')')[1].strip()  # Get the part after the closing parenthesis
        words = part.split()

        words[4] = words[4][:4]  # Extract the first 4 characters of the 5th word (the year)
        year = " ".join(words[:5])
        return year
    except (IndexError):
        return np.nan  # Return NaN if the expected format is not found
```

```python
df['Total Time'] = df['Title'].apply(extract_time)
df.head()
```

```python
from datetime import datetime

def extract_total_months(text):
    if pd.isna(text):
        return np.nan
    try:
        start_str, end_str = text.split(' - ')
        
        start_date = datetime.strptime(start_str, "%b %Y")
        end_date = datetime.strptime(end_str, "%b %Y")

        # Convert year and month to total months
        total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
        return total_months
    except (IndexError, ValueError):
        return np.nan  # Return NaN if the expected format is not found or conversion fails
```

```python
df['Months'] = df['Total Time'].apply(extract_total_months)
df.head()
```
