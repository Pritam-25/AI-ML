### 1. Which Country has the highest population

```python
df[df['population'] == df['population'].max()][['country', 'population']] # Country with the highest population
```

### 2. Top 5 countries with highest democretic score

```python
df.sort_values(by='democracy_score', ascending=False)[['country', 'democracy_score']].head(5) # Top 5 countries by democracy score
```

### 3. How many regions are there

```python
countries_count = df['region'].value_counts() # Count of countries in each region
countries_count
```

```python
total_region = countries_count.count() # Total number of regions
print(f'Total number of regions: {total_region}')
```

### 4. how many countries lie in Eastern Europe region

```python
df[df['region'] == 'Eastern Europe'][['country', 'region']] # Countries in Eastern Europe
```

### 5. who is the political leader of the 2nd highest populated country

```python
df.columns # List of column names
```

```python
# Naive way to get the political leader of the 2nd most populated country

top_2_populated_countries = df.sort_values(by='population', ascending=False)[['country', 'population', 
'political_leader']].head(2)

top_2_populated_countries
```

```python
print("political leader of 2nd most populated country:", top_2_populated_countries.iloc[1]['political_leader'])
```

```python
# âœ… ðŸ”¥ Best Way (Using nlargest())

top2 = df.nlargest(2, 'population')
top2
```

```python
second_most_populated_leader = top2.iloc[1]['political_leader']
print("Political leader of the 2nd most populated country:", second_most_populated_leader)
```

### 8. which country in african continent has highest population


```python
df[df['continent'] == 'Africa'][['country', 'population']].sort_values(by='population', ascending=False).head(5) # Top 5 African countries by population
```

```python
df[df['continent'] == 'Africa'][['country', 'population']].nlargest(1, columns='population') # Countries in Africa with highest population
```
